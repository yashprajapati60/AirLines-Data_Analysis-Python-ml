```python
lis=data["Duration"]
```

```python
new_lis=[]
for i in lis:
    if len(i.split(" "))==1:
        if "m" in  i:
            i="0h " +i
        elif "h" in i:
            i=i+ " 0m"
          
    new_lis.append(i)
```

```python
data["Duration"].unique()
```

```python
data["Duration"]=new_lis
```

```python
data["Duration"]
```

```python
data["du_hr"]=data["Duration"].str.split(" ").str[0].replace("h","",regex=True).astype(int)
data["du_mn"]=data["Duration"].str.split(" ").str[1].replace("m","",regex=True).astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
from sklearn.linear_model import Lasso,Ridge       #Lasso for small data, Ridge for huge data
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
lis=[DecisionTreeRegressor,RandomForestRegressor,KNeighborsRegressor,SVR]
```

```python
for i in lis:
    model=i()
    model.fit(x_train,y_train)
    print(i)
    print(model.score(x_train,y_train))
    print(model.score(x_test,y_test))
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
h=KFold(5)
```

```python
for i in lis:
    print(i)
    print(cross_val_score(i(),x,y,cv=h))
```

```python
for i in lis:
    print(i)
    print(cross_val_score(i(),x,y,cv=h).mean())
```
